```python
#Read in county code data from the Census Bureau into DataFrame
county_codes = pd.read_csv("C:\\Users\\tomas\\OneDrive\\Documents\\county-codes-master.csv", encoding='ISO-8859-1')
county_codes.fips = county_codes.fips.astype(float)

county_codes.head()
```

```python
#Filter county codes so it only includes Pennsylvania
pen_codes = county_codes[county_codes["state_name"] == "Pennsylvania"]
pd.options.mode.chained_assignment = None
pen_codes['county_name'] = pen_codes['county_name'].str.replace('County', '')
pen_codes['county_name'] = pen_codes['county_name'].str.upper()

pen_codes['county_name'] = pen_codes['county_name'].astype(str)

pen_codes.index = pd.RangeIndex(len(pen_codes.index))
pen_codes.head()
```

```python
#Read PA voting data into DataFrame
penvote_data = pd.read_csv("C:\\Users\\tomas\\OneDrive\\Documents\\currentvotestats.csv")

#Drop first row  and empty rows at the end of PA voting data DataFrame
penvote_data = penvote_data.drop(penvote_data.index[0])
penvote_data = penvote_data.drop(penvote_data.index[67:87])

#Drop blank Columns
penvote_data = penvote_data.drop(['Unnamed: 3', 
                                  'Unnamed: 5', 
                                  'Unnamed: 7', 
                                  'Unnamed: 9'], axis=1)

#Label column names
penvote_data.rename( columns={'Information as of 10/19/2020':'name',
                              'Unnamed: 1':'ID', 
                              'Unnamed: 2':'dem_voters', 
                              'Unnamed: 4':'rep_voters', 
                              'Unnamed: 6':'noaff_voters', 
                              'Unnamed: 8':'other_voters', 
                              'Unnamed: 10':'total_voters'}, inplace=True )

pd.to_numeric(penvote_data["dem_voters"])
pd.to_numeric(penvote_data["rep_voters"])
pd.to_numeric(penvote_data["noaff_voters"])
pd.to_numeric(penvote_data["other_voters"])
pd.to_numeric(penvote_data["total_voters"])

penvote_data['name'] = penvote_data['name'].astype(str)

penvote_data.index = pd.RangeIndex(len(penvote_data.index))

penvote_data.head()
```

```python
#Merge PA county codes with voting data
pen_data = pen_codes.join(penvote_data)

pen_data.rename( columns={'fips':'county_code'}, inplace=True )

pen_data.head()
```

```python
#Merge all Dataframes using county codes 
all_data = pd.merge(economic_data, case_data, on="county_code")

all_data = pd.merge(all_data, mask_data, on="county_code")

all_data = pd.merge(all_data, demo_data, on="county_code")

all_data = pd.merge(all_data, social_data, on="county_code")

all_data = all_data.replace('(X)',np.nan)

all_data.head()
```

```python
#Create new column with case rates for each county
all_data["case_rate"] = pd.to_numeric(all_data["10/11/20"]) / pd.to_numeric(all_data["DP05_0001E"])

all_data.head()
```

```python
#Create new column with case rates for each county
all_data["case_rate"] = pd.to_numeric(all_data["10/11/20"]) / pd.to_numeric(all_data["DP05_0001E"])

all_data.head()
```

```python
#Create new column with case rates for each county
all_data["mask_rarely-never"] = pd.to_numeric(all_data["mask_rarely"]) + pd.to_numeric(all_data["mask_never"])

all_data.head()
```

```python
#Add coloumn to specify whether county population is small, medium or large
def label_size (row):
   if pd.to_numeric(row["DP05_0001E"]) < 6000 :
      return 'very small'
   if pd.to_numeric(row["DP05_0001E"]) <  25000:
      return 'small'
   if pd.to_numeric(row["DP05_0001E"]) <  70000:
      return 'medium'
   if pd.to_numeric(row["DP05_0001E"]) <  200000:
      return 'large'
   return 'very large'

all_data.apply(lambda row: label_size(row), axis=1)
all_data["county_size"] = all_data.apply(lambda row: label_size(row), axis=1)

all_data.head()
```

```python
#Filter PA data from all data
allpen_data = all_data[all_data['NAME_x'].str.contains("Pennsylvania")]
allpen_data = pd.merge(allpen_data, pen_data, on='county_code')
allpen_data.head()
```

```python
#Function that specifies whether a county in PA had more democratic or republican voters
def label_party (row):
   if row["dem_voters"] > row["rep_voters"] :
      return 'democratic'
   return 'republican'

allpen_data.apply(lambda row: label_party(row), axis=1)
allpen_data["dom_party"] = allpen_data.apply(lambda row: label_party(row), axis=1)
allpen_data.head()
```

```python
#Linear Regression Model for Case Rate vs Poverty Rate
from scipy.stats import linregress

poverty = pd.to_numeric(all_data["DP03_0119PE"])
case_rate = pd.to_numeric(all_data["case_rate"])

mask = ~np.isnan(poverty) & ~np.isnan(case_rate)
slope, intercept, r_value, p_value, std_err = linregress(poverty[mask], case_rate[mask])
r_value
```

```python
#function that takes in a string which specifies a variable in the all_data DataFrame and creates a facet plot with that log transform of both the variable and coronvirus case rates
def scatter(var):
    all_data['log'+ var] = np.log(pd.to_numeric(all_data[var]))
    all_data['logcase_rate'] = np.log(pd.to_numeric(all_data["case_rate"]))
    
    order = ("very small", "small", "medium", "large", "very large")

    g = sns.FacetGrid(all_data, col="county_size", col_order = order, col_wrap = 5)
    g = g.map(plt.scatter, "log" + var, "logcase_rate")
    
    index = var_names[var_names["Label"]==var].index.values
    
    g.set_axis_labels("log("+ str(var_names.loc[index[0], "Description"]) + ")", "log(Case Rate)")
    g.set_titles(col_template="{col_name} county size")
    g.savefig("facet_plot.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tsmoothie.smoother import *


# set data
x = allpen_data["logDP03_0119PE"]
y = allpen_data["logcase_rate"]

# operate smoothing
smoother = SplineSmoother(n_knots=9, spline_type='cubic_spline')
smoother.smooth(y)

# generate intervals
low, up = smoother.get_intervals('prediction_interval', confidence=0.05)

# plot the first smoothed graph with intervals
plt.figure(figsize=(11,6))
plt.plot(smoother.smooth_data[0], linewidth=3, color='red')
plt.plot(smoother.data[0], '.k')
plt.fill_between(range(len(smoother.data[0])), low[0], up[0], alpha=0.3)
```
